# Task 2: Common bifurcations in nonlinear systems

## Consider a dynamical system on the real line $X = \mathbb{R}$, time $I = \mathbb{R}$, with the evolution described by   $\dot x = \alpha - x^2$. The bifurcation disgram is plotted as follows:


```python
from numpy import *
import matplotlib.pyplot as plt

# prepare plot
fig = plt.figure(figsize=(9,6))
ax1 = fig.add_subplot(1, 1, 1)

def x1(r):
    return sqrt(r)

def x2(r):
    return -sqrt(r)

domain = linspace(0, 1)
ax1.plot(domain, x1(domain), 'b-', label = 'positive steady state', linewidth = 3)
ax1.plot(domain, x2(domain), 'r--', label = 'negative steady state', linewidth = 3)
ax1.legend(loc='upper left')

ax1.plot([0], [0], 'go')
ax1.axis([-0.25, 1, -1.5, 1.5])
ax1.set_xlabel('α')
ax1.set_ylabel('ẋ')
ax1.set_title('Saddle-node bifurcation')
```

## The birufication diagram for the system $\dot x =\alpha -2x^2-3$ is plotted as follows:

```python
from numpy import *
import matplotlib.pyplot as plt

# prepare plot
fig = plt.figure(figsize=(9,6))
ax1 = fig.add_subplot(1, 1, 1)

def xeq1(alpha):
    return sqrt(alpha - 3) / sqrt(2.0)

def xeq2(alpha):
    return -sqrt(alpha - 3) / sqrt(2.0)

domain = linspace(3, 4)
ax1.plot(domain, xeq1(domain), 'b-', label = 'positive steady state', linewidth = 3)
ax1.plot(domain, xeq2(domain), 'r--', label = 'negative state', linewidth = 3)
ax1.legend(loc='upper left')

ax1.plot([0], [0], 'go')
ax1.axis([2.75, 4, -1.5, 1.5])
ax1.set_xlabel('α')
ax1.set_ylabel('ẋ')
ax1.set_title('Saddle-node bifurcation')
```
